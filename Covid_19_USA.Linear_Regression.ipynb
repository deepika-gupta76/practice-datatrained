# Model Training

```python
x=df.drop(columns=['Country_Region','Last_Update','Deaths','Province_State','ISO3'])
x
```

```python
y=df[['Deaths']]
y
```

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

```python
# Importing our model library
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#importing Error Metrices
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#using algorithms via for loop

model=[LinearRegression(),DecisionTreeRegressor(),SVR(),KNeighborsRegressor(),Lasso(),Ridge(),ElasticNet()]

for m in model:
    m.fit(train_x,train_y)
    print('Score of',m,'is:',m.score(train_x,train_y))
    predm=m.predict(test_x)
    print('Error:')
    print('Mean absolute error:',mean_absolute_error(test_y,predm))
    print('Mean square error:',mean_squared_error(test_y,predm))
    print('Root Mean Square Error:',np.sqrt(mean_squared_error(test_y,predm)))
    print("r2_square:",r2_score(test_y,predm))
    print('***********************************************************************************')
    print('\n')
```

observation: From the above models we find that Ridge is working best

```python
#Ridge Model is giving best results,Gridsearch for best results

from sklearn.model_selection import GridSearchCV
ridge=Ridge()
parameters={'alpha':[0.001,0.01,0.1,1],'random_state':range(42,100)}
clf=GridSearchCV(ridge,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
#Using Ridge Model with best results

ridge=Ridge(alpha=1,random_state=42)
ridge.fit(train_x,train_y)
print('Score:',ridge.score(train_x,train_y))
predrd=ridge.predict(test_x)
print('\n')
print('Mean absolute error',mean_absolute_error(test_y,predrd))
print('Mean square error',mean_squared_error(test_y,predrd))
print('Root Mean Square Error',np.sqrt(mean_squared_error(test_y,predrd)))
print("r2_square",r2_score(test_y,predrd)) 
print('\n')
    
```

```python
#cross validate ridge model
from sklearn.model_selection import cross_val_score
score=cross_val_score(ridge,x,y,cv=5,scoring='r2')
print("Score",score)
print('Mean Score:',score.mean())
print("Standard deviation:",score.std())
```
